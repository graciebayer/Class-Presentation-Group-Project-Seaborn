```python
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
countries=pd.read_csv('countries of the world.csv')
```

```python
countries.head()
```

```python
top_5_p=countries.sort_values('Population', ascending=False).head(5)
top_5_p
```

```python
countries.info()
```

```python
top_5=countries.sort_values('Country', ascending=False).head(5)
top_5
```

```python
sns.set(rc={'figure.figsize':(12,10)})
line_plot=sns.lineplot(x='Country',y='Population', data=top_5)
```

```python
bar_plot=sns.barplot(x='Region',y='Area (sq. mi.)',data=top_5)
```

```python
scatter_plot=sns.scatterplot(x='Region', y='Population', data=top_5, color='red')
```

```python
heat_map=top_5[['Population', 'Area (sq. mi.)', 'GDP ($ per capita)']].corr(method='pearson')
cols=['Population', 'Area (sq. mi.)', 'GDP ($ per capita)']
axis=sns.heatmap(heat_map, annot=True, xticklabels=cols, yticklabels=cols)
```
